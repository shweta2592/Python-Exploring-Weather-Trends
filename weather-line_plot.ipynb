```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
temp_df=pd.read_csv("Results.csv")
temp_df.tail()
type(temp_df['MA_10_city'][0])
```

```python
plt.plot(temp_df['year'],temp_df['MA_10_global'],label= 'Global',linewidth=5.0)
plt.plot(temp_df['year'],temp_df['MA_10_city'], label='Singapore',linewidth=5.0)
plt.legend(loc='center right',fontsize=20)
plt.xlabel("Years", fontsize=20)
plt.ylabel("Temperature", fontsize=20)
plt.title('Average Temperature - Singapore vs Globe', fontsize=20)
#plt.rc('xtick', labelsize=20) 
#plt.rc('ytick', labelsize=20) 
plt.xticks(np.arange(min(temp_df['year']), max(temp_df['year'])+10,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='black')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 24
fig_size[1] = 18
plt.rcParams["figure.figsize"] = fig_size
```
